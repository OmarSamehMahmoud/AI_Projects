```python
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Activation
from tensorflow.keras.optimizers import SGD
```

```python
df = pd.read_csv('/content/Boston.csv')

df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
X
```

```python
y
```

```python
model = Sequential()
model.add(Dense(5, input_dim=13, activation='relu'))
model.add(Dense(1))
```

```python
sgd = SGD(0.01)
model.compile(loss='mse',optimizer=sgd)
```

```python
history = model.fit(X,y,epochs=500,verbose=1)
```

```python
pred = model.predict(X)
```

```python
pred
```

```python
plt.scatter(pred, y, c='blue') 
#plt.plot(X.index, pred, color='g') 
plt.show()
```

# Using Sklearn

```python
import numpy as np
import pandas as pd

#Visualization Libraries
import seaborn as sns
import matplotlib.pyplot as plt

#To plot the graph embedded in the notebook
%matplotlib inline

#imports from sklearn library

from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
boston = datasets.load_boston()
```

```python
bos = pd.DataFrame(boston.data, columns = boston.feature_names)
bos['PRICE'] = boston.target

print(bos.head())
```

```python
X = bos.iloc[:, :-1]
y = bos.iloc[:, -1]


X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state=5)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
reg = LinearRegression()
reg.fit(X_train, Y_train)

y_train_predict = reg.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = round(reg.score(X_train, Y_train),2)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
plt.scatter(y_train_predict, Y_train, c='blue') 
#plt.plot(X.index, pred, color='g') 
plt.show()
```

```python
y_test_predict = reg.predict(X_test)

plt.scatter(y_test_predict, Y_test, c='blue') 
plt.show()
```
